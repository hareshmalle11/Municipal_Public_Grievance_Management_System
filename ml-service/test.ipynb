```python
import pandas as pd

# File paths
csv_20k = "grievance_dataset_20k.csv"
csv_50k = "grievance_dataset_50k.csv"
output_csv = "grievance_dataset_70k_merged.csv"

# Load datasets
df_20k = pd.read_csv(csv_20k)
df_50k = pd.read_csv(csv_50k)

# Combine datasets
merged_df = pd.concat([df_20k, df_50k], ignore_index=True)

# Save combined dataset
merged_df.to_csv(output_csv, index=False)

# Display results
print(f"Merged {len(df_20k):,} + {len(df_50k):,} rows = {len(merged_df):,} rows")
print(f"Saved merged file to: {output_csv}")
print(merged_df.head())
```
